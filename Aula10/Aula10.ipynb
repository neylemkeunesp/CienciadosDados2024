```python
df.columns
```

```python
df["winner"] == player
```

```python
df['win'] = df['winner'] == player
df['win'].tail()
```

```python
("{player} has won {vic:.0f}% "
               "of his ATP matches.").format(
                player=player, vic=100*df['win'].mean())
```

```python
df['dblfaults'] = (df['player1 double faults'] /
                   df['player1 total points total'])
```

```python
df['dblfaults'].tail()
```

```python
gb = df.groupby('year')
```
